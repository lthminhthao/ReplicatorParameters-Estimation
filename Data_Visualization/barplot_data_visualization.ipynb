```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pneumococcus_data import data_observation
```

```python
fig, ax = plt.subplots(6, 1, figsize = (20, 35))
fig.subplots_adjust(hspace = 0.4)

Iran, Denmark, Brazil, Nepal, Mozambique = data_observation()
country = [Brazil, Iran, Denmark, Mozambique, Nepal]
label = ['Brazil', 'Iran', 'Denmark', 'Mozambique', 'Nepal']

for i in range (0, len(label)):
    sns.barplot(ax = ax[i], x = country[i].strain, y = country[i].cases, alpha = .5, label = '{}'.format(label[i]), color = 'blue')
    ax[i].set_xlabel('Serotype', fontsize = 20)
    ax[i].set_ylabel('Serotype frequency', fontsize = 20)
    ax[i].tick_params(axis='x', which='major', labelsize = 14, rotation = 45)
    ax[i].tick_params(axis='y', which='major', labelsize = 14)
    ax[i].legend(fontsize=22)
    
color = ['red', 'lime', 'fuchsia', 'aqua', 'mediumblue']
for i in range (0, len(country)):
    ax[-1].plot(np.linspace(1, len(country[i]), len(country[i])), country[i].cases, linewidth = 4.5, color = '{}'.format(color[i]), label = '{}'.format(label[i]))

ax[-1].set_xlabel('Serotype rank', fontsize = 20)
ax[-1].set_ylabel('Serotype frequency', fontsize = 20)
ax[-1].tick_params(axis='x', which='major', labelsize = 14)
ax[-1].legend(fontsize=16)

plt.rcParams["figure.dpi"] = 300
plt.show()
```
